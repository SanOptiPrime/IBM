CHECK FEATURES THAT COULD HAVE OUTLIERS:
(i) Start station latitude 
(ii) Start station longitude 
(iii) End station latitude 
(iv) End station longitude 
(v) Birth year

```python
def find_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers
```

```python
outliers = find_outliers_IQR(df["birth year"])

print("number of outliers: " + str(len(outliers)))

print("max outlier value: " + str(outliers.max()))
print("min outlier value: " + str(outliers.min()))
```

**It is more convenient to create separate columns - to make use of the starting time and stopping time better. **

The following steps are carried out: <br>
1. Hour column introduced to find out the most active hours of bike usage <br>
2. Age column introduced to carry out further analysis on influence of different age groups <br>
3. Age_Group - based on the age, they fall into different age groups <br>


```python
temp = edadf
```

Hour column introduction

```python
temp["starttime"] = pd.to_datetime(temp["starttime"])
```

```python
temp["stoptime"] = pd.to_datetime(temp["stoptime"])
```

```python
temp.info()
```

```python
temp["Hour"] = temp["starttime"].dt.hour
```

```python
temp.head()
```

```python
temp.shape
```

Age column introduced

```python
temp['Age'] = 2022 - temp['birth year'] 
```

```python
temp.head()
```

Assigning Age groups to the Bike users

```python
Age_Groups = ["<20", "20-29", "30-39", "40-49", "50-59", "60+"]
Age_Groups_Limits = [0, 20, 30, 40, 50, 60, np.inf]
Age_Min = 0
Age_Max = 100
temp["Age_group"] = pd.cut(temp["Age"], Age_Groups_Limits, labels=Age_Groups)
```

```python
temp
```

```python
temp.info()
```

Gender Based Variation among data <br>
1 : Male, 2 : Female, 0 : Unknown

```python
import plotly.express as px
px.pie(values = temp['gender'].value_counts(),
       names =temp['gender'].value_counts().index,
       title ="Gender Variation")
```

Peak hours of Bike Usage

```python
px.bar(x=temp["Hour"].value_counts().index,
       y=temp["Hour"].value_counts().values,
       title = "Hour usage of Citi Bikes",
       labels={'x':'Time',"y":"Number of people using bike"})
```

Most frequently and Least frequently used Start and End stations 

```python
#Start Stations
px.bar(x=temp["start station name"].value_counts().index,
       y=temp["start station name"].value_counts().values,
       labels={'x':'Start Station Name',"y":"Count"})
```

```python
#End Stations
px.bar(x=temp["end station name"].value_counts().index,
       y=temp["end station name"].value_counts().values,
       labels={'x':'End Station Name',"y":"Count"})
```

Age Distribution



```python
plt.figure(figsize = (20,5))
sns.countplot(temp["Age_group"],palette = "Set1")
plt.show()
```

Most popular 50 routes travelled using Citi bike

```python
trips_df = pd.DataFrame()
trips_df = temp.groupby(['start station name','end station name']).size().reset_index(name = 'Number of Trips')
trips_df = trips_df.sort_values('Number of Trips',ascending = False)
trips_df["start station name"] = trips_df["start station name"].astype(str)
trips_df["end station name"] = trips_df["end station name"].astype(str)
trips_df["Routes"] = trips_df["start station name"] + " to " + trips_df["end station name"]
trips_df = trips_df[:50]
trips_df = trips_df.reset_index()
trips_df
```
